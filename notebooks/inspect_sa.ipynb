```python
import json
from collections import Counter

from datasets import load_dataset
from rich.console import Console
from rich.panel import Panel

console = Console()
```

```python
split_data = load_dataset(
    "codeparrot/codeparrot-clean-valid",
    split="train",
    cache_dir="/Users/chenghao/Downloads/Projects/text-dedup/.cache"
)
```

```python
results = json.load(open("/Users/chenghao/Downloads/Projects/text-dedup/sa_results.json"))
```

```python
from collections import defaultdict

duplication_ratios = {}
snippets = defaultdict(set)
for res in results:
    doc = split_data[res["query_index"]]
    doc_text = doc["content"].encode("utf-8")
    total_size = len(doc_text)
    dup_size = 0
    if not res["byte_slices"]:
        continue
    for slice in res["byte_slices"]:
        dup_size += slice[1] - slice[0]
        snippets[doc_text[slice[0]:slice[1]].decode("utf-8", errors="ignore")].add(res["query_index"])
    if dup_size == 0:
        continue
    duplication_ratios[res["query_index"]] = dup_size / total_size
```

```python
min(map(len, snippets.values()))
```

```python
minhash_results = json.load(open("/Users/chenghao/Downloads/Projects/text-dedup/minhash_results.json"))
```

```python
minhash_ids = set()
for res in minhash_results:
    minhash_ids.add(res["query_index"])
    minhash_ids.update(res["references"])
```

```python
len(minhash_ids)
```

```python
min(minhash_ids, key=lambda x: duplication_ratios.get(x, float('inf')))
```

```python
len([x for x in duplication_ratios if duplication_ratios[x] > 0.9])
```

```python
for x in duplication_ratios:
    if duplication_ratios[x] > 0.8 and x not in minhash_ids:
        print(x, duplication_ratios[x])
```
